```python
original_orbits = pd.read_csv("../skyMapper_main_folder/sbdb_query_results.csv",low_memory=False,float_precision = "round_trip")
```

```python
orbits = original_orbits
prbits = orbits[(orbits["H"] >= 18) & (orbits["H"] <= 20)]
orbits = orbits.head(5000)

olist = []

for index, row in orbits.iterrows():
    if row["H"] is not None:
        H = row["H"]
    else:
        H = 20.0
    if row["G"] is not None:
        G = row["G"]
    else:
        G = 0.15
        
    epoch = Time(row["epoch_mjd"], scale="tdb", format="mjd")
    epoch_tt_mjd = epoch.tt.mjd
    epoch_tt_mjd = epoch.tt.mjd
    orbit = Orbit.keplerian(index, row["a"], row["e"], row["i"], row["om"], row["w"], row["ma"], epoch_tt_mjd, EpochTimescale.TT,H, G)    
    olist.append(orbit)
```
