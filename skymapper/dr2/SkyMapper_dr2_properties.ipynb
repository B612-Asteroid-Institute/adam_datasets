```python
import pyvo as vo
import pandas as pd
tap_service = vo.dal.TAPService("https://api.skymapper.nci.org.au/public/tap/")   
```

**There are NUll obj_id in dr2.photometry**

```python
query = f"""SELECT * FROM dr2.photometry WHERE object_id IS NULL"""
tap_results = tap_service.search(query, maxrec=5)
pd.DataFrame(tap_results) 
```

**There are Null magnitude for nonNULL object_id in dr2.photometry**

```python
query = f"""SELECT * FROM dr2.photometry WHERE object_id is NOT NULL AND mag_psf is NULL"""
tap_results = tap_service.search(query, maxrec=5)
pd.DataFrame(tap_results) 
```
