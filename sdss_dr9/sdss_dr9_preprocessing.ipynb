```python
col = raw_data.columns.tolist()
ids = [item for item in col if 'TAI' in item]
ids
```

```python
'''
columns needed:
obs_id : program assigned unique observation ID (objID)
exposure_id : program assigned exposure ID (a single exposure can have many observations) (fieldID)
mjd_utc : Modified Julian Date [MJD] in UTC (convert TAI_i to mjd)
ra : J2000 Right Ascension in degrees (ra)
dec : J2000 Declination in degrees (dec)
ra_sigma : 1-sigma uncertainty in RA in degrees (raErr)
dec_sigma : 1-sigma uncertainty in Dec in degrees (decErr)
filter : band in which observation was made (i)
mag : magnitude in filter (i)
mag_sigma : 1-sigma magnitude uncertainty in magnitude (err_i)
obscode : MPC observatory code (645)


'''
```

```python
raw_data['TAI_i']
# use TAI_i as time of observation (units in sec/TimeUnixTai)
# convert to mjd via astropy
```

```python
raw_data['mjd_i'] = raw_data['TAI_i']/(24*3600)
```

```python
#times = Time(raw_data['TAI_i'], format='unix_tai')
#times.mjd
```

```python
formatted_data = {
    'obs_id':raw_data['objID'],
    'exposure_id':raw_data['fieldID'],
    'mjd_utc':raw_data['mjd_i'],
    'ra':raw_data['ra'],
    'dec':raw_data['dec'],
    'ra_sigma':raw_data['raErr'],
    'dec_sigma':raw_data['decErr'],
    'filter':'i',
    'mag':raw_data['i'],
    'mag_sigma':raw_data['err_i'],
    'obscode':645
}

formatted_df = pd.DataFrame(formatted_data)
formatted_df
```
